```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress
```

```python
df = pd.read_csv('/content/drive/MyDrive/Programação/Curso analise de dados do FreeCodeCamp/epa-sea-level.csv')
```

```python
plt.figure(figsize=(14, 7))
plt.scatter(y='CSIRO Adjusted Sea Level', x='Year', data=df)
plt.yticks(ticks=range(-1, 11), labels=range(-1, 11))
plt.title('Rise in Sea Level')
plt.ylabel('Sea Level (inches)')
plt.xlabel('Year')
plt.show()
```

```python
x1 = range(df["Year"].iloc[0], 2051)
angular, linear, r, p, se = linregress(df["Year"], df["CSIRO Adjusted Sea Level"])
plt.plot(x1, (angular*x1 + linear))
plt.title('Rise in Sea Level')
plt.ylabel('Sea Level (inches)')
plt.xlabel('Year')
plt.show()
```

```python
x1 = range(2000, 2051)
df_2000 = df[df['Year'] >= 2000]
angular, linear, r, p, se = linregress(df_2000["Year"], df_2000["CSIRO Adjusted Sea Level"])
plt.plot(x1, (angular*x1 + linear))
plt.title('Rise in Sea Level')
plt.ylabel('Sea Level (inches)')
plt.xlabel('Year')
plt.show()
```
